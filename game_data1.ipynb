We used a game sales data sheet that listed the sot selling PS4 games that came out 2018 and before. We spesified the top 99 in our analysis, and chose to compare how well each "Genre" sold. We thought we could use this data to help cater out website. I had also attempted to compare sales in each region according to genre, but failed to find a way to do it without making the calculations and manually putting in the numbers for each genre.
```python
import tensorflow as tf
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import csv
import re
from operator import itemgetter
from tensorflow.keras import layers
```

```python
df = pd.read_csv('./static/data/PS4_GamesSales.csv')
df1 = df.copy()
```

```python
group1 = df1.groupby('Genre')['Global'].sum()

group1 = group1.sort_values(ascending=False)
```

```python
group1.plot(kind="barh", figsize=(25,10))
plt.title("Global Sales of Top 100 Video Games, in Mil/units", fontdict={"fontsize":30, "weight":"bold"}, pad=20)
plt.show()
```
